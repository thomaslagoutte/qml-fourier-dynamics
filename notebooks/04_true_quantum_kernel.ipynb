```python
import numpy as np
from quantum_learning_dynamics import Experiment
from quantum_learning_dynamics.hamiltonians import InhomogeneousTFIM
from quantum_learning_dynamics.observables import LocalZ

# Setup a d>1 model (which previously took 22+ minutes for features)
model = InhomogeneousTFIM(num_qubits=2)
observable = LocalZ(num_qubits=2, site=0)

exp_kernel = Experiment(
    model=model,
    observable=observable,
    method="true_quantum_kernel", # THE NEW ROUTE
    tau=1.0,
    r_steps=2,
    execution_mode="hardware",
    shots=4000, 
    seed=42
)

# Train on just 15 graphs, test on 5 (This requires only 15x15 + 15x5 = 300 circuits total!)
# Compare this to the (4*2 + 1)^3 = 729 circuits PER GRAPH (14,580 total) the old way required.
result = exp_kernel.run(num_train=15, num_test=5)

print(f"--- True Quantum Kernel Results ---")
print(f"Trotter MSE: {result.mse_trotter:.6f}")
print(f"Exact MSE:   {result.mse_exact:.6f}")
```
